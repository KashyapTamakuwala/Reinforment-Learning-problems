```python
from Gridworld import Gridworld , printPolicy , printV
from policyEvaluation import evaluatePolicy
from policyImprovement import improvePolicy
from iteratevalue import iterateValues
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#defining Wall
wall=[]
for i in range(4,15):
    wall.append((12,i))
for i in range(7,12):
    wall.append((i,2))
for i in range(3,13):
    wall.append((7,i))
for i in range(0,11):
    wall.append((3,i))

print(wall)
```

```python
alpha=[0.2,0.4]
m=15
n=15
#wall=[(2,i) for i in range(0,3)]
x=Gridworld(m,n,wall,alpha[0])
x.renderHeatMap()
```

```python
Gamma = 1
Theta = 1e-6
stable = False

V = {}
for state in x.statespaceplus:
    V[state] = 0

policy = {}
for state in x.statespace:
    policy[state] = x.possibleaction

```

```python
k = 0
while True:
    V = evaluatePolicy(x, V, policy, Gamma, Theta)
    array=printV(V, x)
    stable,policy = improvePolicy(x,V,policy,Gamma)
    print("stable",stable)
    k = k+1
    if stable != False:
        break

print("while loop iteration : ",0)

poli=printPolicy(policy,x)
```

```python
from tabulate import tabulate
table = tabulate(np.array(poli, dtype='object'), tablefmt="fancy_grid")

# output
print(table)
```
